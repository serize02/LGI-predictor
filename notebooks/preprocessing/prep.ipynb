```python
df = pd.read_csv('data/raw/data.csv')
df.head()
```

```python
drops = ['Unnamed: 0', 'EDAD-R', 'Diagnóstico']
df.drop(columns=drops, axis=1, inplace=True)
```

```python
df.columns
```

```python
df['Complicaciones'].unique()
```

```python
df.fillna('-', inplace=True)
df['Complicaciones'] = df['Complicaciones'].apply(lambda x: x.capitalize().split(','))
```

```python
st = set()

def add_to_set(x: list):

    for i in x:
        st.add(i)

df['Complicaciones'].apply(add_to_set)
```

```python
ls = list(st)

for i in ls:
    df[i] = df['Complicaciones'].apply(lambda x: i in x)

df[ls]
```

```python
df.drop(columns=['Complicaciones', '-'], axis=1, inplace=True)
```
